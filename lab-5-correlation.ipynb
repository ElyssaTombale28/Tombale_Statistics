```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#load dataset
df = pd.read_csv(r'dataset/current-power.csv', 
                    delimiter=",")
df
```

```python
#covariance of the current and power
df.cov()
```

```python
#scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs. Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
# correlation coefficient of current and power
df.corr()
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='#1C1F24',
            alpha=0.6,
            label='Data Points')

#line plot
plt.plot(x, line,
         color='#126782',
         label=f'Regression line (r = {r_value:.3f})')

plt.title('Device Current vs. Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()


```

Air Quality

```python
#Load dataset
air = pd.read_csv(r"dataset\air-quality.csv",
                    delimiter=",")
air
```

```python
#covariance of the air
air.cov()
```

```python
# correlation coefficient of current and power
air.corr()
```

```python
#regression line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='#1C1F24',
            alpha=0.6,
            label='Data Points')

#line plot
plt.plot(x, line,
         color='#126782',
         label=f'Regression line (r = {r_value:.3f})')

plt.title('NOx vs. Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()


```
