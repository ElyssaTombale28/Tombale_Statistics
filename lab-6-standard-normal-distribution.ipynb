```python
#imports and packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Load Dataset
df = pd.read_csv(r"dataset\current-test.csv",
                    delimiter=',',
                    index_col='Sample')
df
```

```python
df['Current'][2]
```

```python
#summary of statistics
df.describe()
```

```python
#Z-scores
df['Z-score'] = stats.zscore(df['Current'], ddof=1)
df
```

```python
#Z-scores summary of statistics
df['Z-score'].describe()
```

Histogram

```python
#histogram plot
plt.hist(df['Current'], bins=5,
         alpha= 0.8,
         color= '#fa6194ff' ,
         edgecolor= "#f0f0f0ff")

plt.title('Histogram of Current Values',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

Normal Distribution

```python
#normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

#x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color= '#fa6194ff',
        linewidth=2)

plt.title('Normal Distribution of Current Values',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
plt.hist(df['Current'], bins=5,
         density= True,
         alpha= 0.8,
         color= '#fa6194ff' ,
         edgecolor= "#f0f0f0ff")

plt.title('Normal Distribution of Current Values',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')

#normal distribution curve
p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
        color= "#000000ff",
        linewidth=2)
plt.show()
```

Standard Normal Distribution N(0, 1)

```python
#standard normal distribution N(0,1)
#x-axis
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()
x = np.linspace(x_min, x_max, 100)

p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof= 1))

plt.plot(x, p,
        color= "#fa6194ff",
        linewidth=2)

plt.title('Standard Normal Distribution of Current Values N(0, 1)',
          font='Palatino Linotype',
          fontsize=12)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```
